```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
observed = np.load("observed_data/nlinear_no_smoothing/nlinear_no_smoothing.npy")
observed_linear_smooth = np.load("observed_data/nlinear_smoothing/nlinear_smoothing.npy")
```

```python
prey = observed[:,0]
predator = observed[:,1]
prey_smooth = observed_linear_smooth[:,0]
predator_smooth = observed_linear_smooth[:,1]
t = np.linspace(0, 10, 100)
plt.title("Observed data with $\epsilon\sim(0,t^2)$")
plt.scatter(t, prey, label="Prey")
plt.scatter(t, predator, label="Predator")
plt.plot(t, prey_smooth, label="Prey Spline")
plt.plot(t, predator_smooth, label="Prey Spline")
plt.xlabel("t")
plt.ylabel("Population")
plt.legend()
# plt.savefig("plots/nlinear_no_smoothing/observed_data.png")
plt.show()
```

```python
observed_no_smooth = np.load("observed_data/n0.5_no_smoothing/n0.5_no_smoothing.npy")
observed_smooth = np.load("observed_data/n0.5_smoothing/n0.5_smoothing.npy")
```

```python
prey_no_smooth = observed_no_smooth[:,0]
predator_no_smooth = observed_no_smooth[:,1]
prey_smooth = observed_smooth[:,0]
predator_smooth = observed_smooth[:,1]
t = np.linspace(0, 10, 100)
plt.title("Observed data with $\epsilon\sim(0,0.5^2)$")
plt.scatter(t, prey_no_smooth, label="Prey")
plt.scatter(t, predator_no_smooth, label="Predator")
plt.plot(t, prey_smooth, label="Prey Spline")
plt.plot(t, predator_smooth, label="Predator Spline")
plt.xlabel("t")
plt.ylabel("Population")
plt.legend()
plt.tight_layout()
plt.savefig("plots/n0.5_no_smoothing/observed_w_wout_smoothing.png")
plt.show()
```
