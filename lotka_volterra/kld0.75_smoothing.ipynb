## Observed Data with Simulated Data using KLD

```python
def abc_posterior_data(nparams: int, distances: np.ndarray, distance_quantile: float, distance_metric: str) -> np.ndarray:
    if distance_metric == "Wasserstein Distance":
        index = 2
    elif distance_metric == "Energy Distance":
        index = 3
    elif distance_metric == "Maximum Mean Discrepancy":
        index = 4
    elif distance_metric == "Cramer-von Mises Distance":
        index = 5
    elif distance_metric == "Kullback-Leibler Divergence":
        index = 6

    # Calculate quantile from given quantile
    threshold = np.nanquantile(distances[:,index], distance_quantile)
    ## Identify Alpha and Beta after filtering
    posterior_params = distances[distances[:,index] <= threshold][:,0:nparams]
    posterior_kld = distances[distances[:,index] <= threshold][:,6]

    return posterior_params, np.column_stack((posterior_params, posterior_kld))
```

```python
run = np.load("runs/n0.75_smoothing/run1.npy")
```

```python
kld_low = abc_posterior_data(2, run, 0.001, "Kullback-Leibler Divergence")[0]
alpha_median, beta_median = np.median(kld_low[:,0]), np.median(kld_low[:,1])

def dUdt(t, state, theta_a, theta_b):
    x, y = state[..., 0], state[..., 1]
    dxdt = theta_a*x - x*y
    dydt = theta_b*x*y - y
    return torch.stack((dxdt, dydt), dim=-1)

ic = torch.full((1, 2), 0.5).cuda()
t = torch.linspace(0, 10, 100).cuda()
sim_sol = odeint(lambda t, state: dUdt(t, state, alpha_median, beta_median), ic, t, method='rk4')
prey_sol = np.nan_to_num(np.array(sim_sol.cpu()[:, :, 0]))
predator_sol = np.nan_to_num(np.array(sim_sol.cpu()[:, :, 1]))
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, observed_smoothing[:,0], label="Prey")
plt.plot(x, observed_smoothing[:,1], label="Predator")
plt.scatter(x, observed_no_smoothing[:,0])
plt.scatter(x, observed_no_smoothing[:,1])
plt.plot(x, prey_sol, label="Simulated Prey")
plt.plot(x, predator_sol, label="Simulated Predator")
plt.legend()
plt.xlabel("t")
plt.ylabel("Population")
plt.title("$\epsilon\sim N(0, 0.75^2)$ KLD")
# plt.savefig("plots/n0.75_no_smoothing/sim_kld_sol.png")
plt.show()
```

## Observed Data with Simulated Data from Other Metrics

```python
other_metrics = ["Wasserstein Distance", "Energy Distance", "Maximum Mean Discrepancy", "Cramer-von Mises Distance"]

for metric in other_metrics:
    temp_data = abc_posterior_data(2, run, 0.001, metric)[0]
    alpha_median, beta_median = np.median(temp_data[:,0]), np.median(temp_data[:,1])
    sim_sol = odeint(lambda t, state: dUdt(t, state, alpha_median, beta_median), ic, t, method='rk4')
    prey_sol = np.nan_to_num(np.array(sim_sol.cpu()[:, :, 0]))
    predator_sol = np.nan_to_num(np.array(sim_sol.cpu()[:, :, 1]))

    x = np.linspace(0, 10, 100)
    plt.plot(x, observed_smoothing[:,0], label="Prey")
    plt.plot(x, observed_smoothing[:,1], label="Predator")
    plt.scatter(x, observed_no_smoothing[:,0])
    plt.scatter(x, observed_no_smoothing[:,1])
    plt.plot(x, prey_sol, label="Simulated Prey")
    plt.plot(x, predator_sol, label="Simulated Predator")
    plt.legend()
    plt.xlabel("t")
    plt.ylabel("Population")
    plt.title(f"$\epsilon\sim N(0, 0.75^2)$ {metric}")
    if metric == "Wasserstein Distance":
        save_name = "wd"
    elif metric == "Energy Distance":
        save_name = "ed"
    elif metric == "Maximum Mean Discrepancy":
        save_name = "mmd"
    elif metric == "Cramer-von Mises Distance":
        save_name = "cvmd"
    # plt.savefig(f"plots/n0.75_no_smoothing/sim_{save_name}_sol.png")
    plt.show()
```
